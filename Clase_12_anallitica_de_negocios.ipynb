```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf
### Tensor flow me permite usar las caracteristicas de las redes neuronales de tensor flow-google
import keras
### Libreria de paquetes que me permite programar la red neuronal que quiero
from keras.models import Sequential
### Esta es el modelo de activacion de capas en secuencia
from keras.layers import Dense
### Esto me permite densificar la capa de la red nuronal
### Es decir cuantas neuronas quiero en cada capa
```

```python
### Montar la base de datos
ruta='/content/Base de datos esg.xlsx'
base=pd.read_excel(ruta)
base.head()


base2=base.dropna()
base2.columns
```

```python
### Preparacion de los datos
y=base2[['ESG Score ']]
x=base2[['Environmental Score ','Social Score ', 'Governance Score ',
         'Board Gender Diversity Score ','CEO Board Member ', 'Beta ', 'WACC ']]

s_scaler=StandardScaler()
y_s=s_scaler.fit_transform(y)
x_s=s_scaler.fit_transform(x)

x_train,x_test,y_train,y_test=train_test_split(x_s,y_s,train_size=.80,random_state=42)
```

```python
##Vamos a montar el modelo adaline para predecir ESG score a partir de las variables en x

modelA=Sequential()
modelA.add(Dense(1,activation='linear', input_dim=7,use_bias=False))##
modelA.compile(loss='mean_squared_error',optimizer='adam')
historyA=modelA.fit(x_train,y_train,epochs=150,batch_size=32)
ypA=modelA.predict(x_test)
```

```python
###Voy a hacer graficas para ver como se comporto el modelo

##Grafica de dispersiГіn
plt.figure()
plt.scatter(y_test,ypA)
plt.xlabel('y_test')
plt.ylabel('ypA')
plt.show()

##Grafica de distribucion
plt.figure()
sns.displot((y_test.flatten(),ypA.flatten()),kde=True)
plt.xlabel('y_test')
plt.ylabel('ypA')
plt.show()

##correlaciГіn
print(np.corrcoef(y_test.flatten(),ypA.flatten()))
```

```python
### Voy a buscar los pesos del modelo
wA=modelA.layers[0].get_weights()[0]
wA
```

```python
##Voy a montar el modelo MADALINE multiples capas multiples neuronas

modelM=Sequential()
modelM.add(Dense(4,activation='linear', input_dim=7,use_bias=False))
modelM.add(Dense(1,activation='relu',use_bias=False))
modelM.compile(loss='mean_squared_error',optimizer='adam')
historyM=modelM.fit(x_train,y_train,epochs=150,batch_size=32)
ypM=modelM.predict(x_test)
```

```python
## ver los pesos

w0M=modelM.layers[0].get_weights()[0]
w1M=modelM.layers[1].get_weights()[0]
w0M,w1M
wtM=np.dot(w0M,w1M)
wtM
#
```
