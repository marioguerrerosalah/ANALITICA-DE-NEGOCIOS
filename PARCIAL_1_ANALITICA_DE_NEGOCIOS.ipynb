```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
ruta="/content/base de datos salida de clientes bancos.xlsx"
base=pd.read_excel(ruta)
```

```python
base.shape # la base de datos tiene una dimension de 10000 filas y 13 columnas
```

BORRAR E IDENTIFICAR VACIOS Y DUPLICADOS

```python
base.isnull().sum()
base2=base.dropna()

base.duplicated().sum()
base2.duplicated().sum()
```

```python
base2['Geography'].value_counts()
base2['NumOfProducts'].value_counts()
base2['HasCrCard'].value_counts()
base2['IsActiveMember'].value_counts()
base2['Exited'].value_counts()
```

```python
#CreditScore
base2["CreditScore"].mean()
base2["CreditScore"].std()
base2["CreditScore"].min()
base2["CreditScore"].max()

#Balance
base2["Balance"].mean()
base2["Balance"].std()
base2["Balance"].min()
base2["Balance"].max()

#EstimatedSalary
base2["EstimatedSalary"].mean()
base2["EstimatedSalary"].std()
base2["EstimatedSalary"].min()
base2["EstimatedSalary"].max()

```

```python
base2= pd.get_dummies(base2, columns=['Geography', 'Gender'], drop_first=True)
```

```python
X = base2.drop(columns=['Exited', 'CustomerId', 'Surname'])  # Variables predictoras
y = base2['Exited']  # Variable objetivo
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#REGRESION LOGISTICA
##definir el modelo
logit=LogisticRegression()

###entrenar el modelo
logit.fit(X_train,y_train)

##Prediccion por fuera de la muestra
y_pred=logit.predict(X_test)

###MATRIZ DE CONFUSION

mc_lr =confusion_matrix(y_test,y_pred)
print(confusion_matrix(y_test,y_pred))

#### resumen de las metricas
print(classification_report(y_test,y_pred))
```

```python
##NAIVE BAYES


##definir el modelo
nb=GaussianNB() ##este es el modelo de naive bayes

###entrenar el modelo
nb.fit(X_train,y_train) # Changed x_train to X_train

##Prediccion por fuera de la muestra
y_pred=nb.predict(X_test)

##MATRIZ DE CONFUSION
mc_nb=confusion_matrix(y_test,y_pred)
print(confusion_matrix(y_test,y_pred))

###reporte de la clasficación
print(classification_report(y_test,y_pred))
```
